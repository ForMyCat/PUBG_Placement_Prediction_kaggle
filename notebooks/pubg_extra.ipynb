```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
feature_importance_dict = {'walkDistance' : 0.7637463,
'killPlace' : 0.13593936,
'boosts' : 0.040704165,
'weaponsAcquired' : 0.00601034,
'damageDealt' : 0.00072893145,
'heals' : 0.002844148,
'kills' : 0.03896579,
'longestKill' : 0.005002158,
'killStreaks' : 0.0060586906}
```

```python
feature_importance_pd = pd.DataFrame({'keys':feature_importance_dict.keys(),'values':feature_importance_dict.values()})
feature_importance_pd = feature_importance_pd.sort_values(by='values',ascending=False)
```

```python
plt.figure(figsize=(8, 10))
sns.barplot(data=feature_importance_pd,y='keys', x='values', palette='Set2')
plt.ylabel("Feature")
plt.ylabel("Importance")
plt.title('Most Important Features - XGBoost gbtree')
```

```python
features = ['walkDistance',
'killPlace',
'boosts',
'weaponsAcquired',
'damageDealt',
'heals',
'kills',
'longestKill',
'killStreaks']
```

```python
feature_coefficient_pd = pd.DataFrame({'feature':features,'coef':[ 0.0753954,-0.056491,0.0386121, 0.041611,0.00856005,0.0182801,0.00325765,0.0125268,-0.00086056]})
feature_coefficient_pd = feature_coefficient_pd.sort_values(by='coef',ascending=False)
```

```python
plt.figure(figsize=(8, 10))
sns.barplot(data=feature_coefficient_pd,y='feature', x='coef', palette='Set2')
plt.ylabel("Feature")
plt.ylabel("Coefficient")
plt.title('Features Correlation - XGBoost gblinear')
```
